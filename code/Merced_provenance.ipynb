## Geologic background

### Interpretation of provenance
Hall (1965) identified a change in provenance within the Merced Fm from local Franciscan sources to sediment sourced from the Sacramento and San Joaquin River systems, based on mineralogical differences. Hall places the provenance change ~80 m below a distinctive clay-rich shell bed named the "Upper Gastropod Bed" by Ashley (1895). This is in agreement with Hunter and Clifton (1987), who have the Upper Gastropod Bed at 210 m in their section (top of sequence R1) and the provenance change at 290 m (boundary between sequences P and Q). 

The heavy mineral assemblage in the lower Merced is dominated by hornblende, epidote, tremolite-actinolite, and glaucophane. It also contains distinctive Franciscan minerals such as pumpellyite and lawsonite. These minerals become rare in the upper Merced, which is characterized by abundant hypersthene, hornblende, and augite. These minerals are interpreted as coming from an andesitic source. There is also a change in feldspar composition within the Merced Fm, with Na-rich plagioclase being common in the lower Merced and Ca-rich plagioclase being more abundant in the upper Merced.

In addition to samples from the Merced Fm, Hall (1965) also studied the heavy mineral composition of samples from three potential sediment sources: San Francisco Bay, the Sonoma Volcanics, and the Central Valley. The San Francisco Bay samples are sands from drill cores, while the Sonoma Volcanics and Central Valley samples are river sands. Through comparison with these modern sources, Hall (1965) was able to identify rocks of the Franciscan complex as the sediment source for the lower Merced and Sierran detritus from the Sacramento/San Joaquin Rivers as the sediment source for the upper Merced.


This provenance change is important because it is interpreted as marking the establishment of Central Valley drainage into the San Francisco Bay via the Sacramento and San Joaquin Rivers. Prior to this, the Central Valley was a closed basin and contained a large freshwater lake, Lake Clyde (sometimes also referred to as Lake Corcoran).

### Lake Clyde and the Corcoran Clay
At its greatest extent, Lake Clyde occupied the entire San Joaquin Valley and parts of the Sacramento Valley. The lake is represented in the geologic record by the Corcoran Clay member of the Turlock Lake Formation. 

According to Sarna-Wojcicki (2021), the Corcoran Clay *"represents an estimated time interval of 0.77 m.y. (ca. 1.4–0.63 Ma), based on magnetostratigraphy, tephrochronology, and stratigraphic thickness extrapolations"* but how exactly the 1.4 Ma age of formation was determined is unclear.

The best chronostratigraphic constraints on the Corcoran Clay come from a drill core near Wasco, in the San Joaquin Valley (Davis et al., 1977; Sarna-Wojcicki, 2021). The Corcoran Clay at this site is 12-14 m thick, and represents the late history of Lake Clyde. The clay layer contains the Bishop Tuff (764.8 ± 0.6 ka; Andersen et al., 2017) near its base. The Brunhes-Matuyama reversal (ca. 773 ka; Gibbard and Head, 2020) was also identified at the base of the clay, ~4 m below the Bishop Tuff (Davis et al., 1977). The Lava Creek B ash bed (ca. 631 ka; Matthews et al., 2015) is located at the top of the clay layer and represents the best upper age constraint on the Corcoran Clay.

Sarna-Wojcicki (2021) proposes that Central Valley drainage into the San Francisco Bay was initiated by the spillover of Lake Clyde through the Carquinez Strait towards the end of MIS 16, ca. 630 ka. This spillover event was likely caused by increased influx of meltwater due to the larger-amplitude climate fluctuations associated with the mid-Pleistocene transition. The newly established drainage system would have flowed through the Colma Gap south of the Golden Gate and entered the Merced Embayment to the west. The provenance change in the Merced Fm identified by Hall (1965) represents this drainage reorganization.

### Chronostratigraphic constraints on the provenance change
The only well-dated bed in the Merced Fm is the Rockland Ash, located near the top of sequence S2 in Clifton and Hunter's framework (170 m below the top of the section; Clifton and Hunter, 1987). The provenance change is located 290 m below the top of the section (Clifton and Hunter, 1987), meaning the Rockland Ash and the provenance change are separated by 120 m of strata.

The Rockland Ash has an age of ca. 600 ka based on the following two studies:
- Lanphere et al. (2004): 40Ar/39Ar plagioclase age of 609 ± 7 ka
- Coble et al. (2017): U/Pb zircon age of 598 ± 13 ka; (U-Th)/He zircon age of 599 ± 12 ka

Given the presence of the Lava Creek B Tuff near the top of the Corcoran Clay and the presence of the Rockland Ash in the Merced Fm above the provenance change, the timing of the outflow of Lake Clyde and subsequent reorganization of the Central Valley drainage is constrained to between 0.6 and 0.63 Ma.

Sarna-Wojcicki (2021) also reports reworked, disseminated glass shards of the Bishop Tuff within the Merced Fm, from just above the provenance change. These shards are interpreted as being reworked from Bishop Tuff deposits within the Corcoran Clay following the spillover of Lake Clyde.

### Implied sedimentation rates
There are four potentially chronostratigraphically useful horizons currently known within the Merced Fm:
1. The base of the formation: Ingram and Ingle (1998) developed two age models based on strontium isotope stratigraphy. Model 1 has a basal age of 2.4 Ma, and Model 2 has a basal age of 4.8 Ma. Our best fit linear age model which incorporates the most recent age of the Rockland Ash gives a basal age of 3.3 Ma (see the Merced_Strontium notebook).
2. The provenance change: As detailed above, the provenance change is constrained between 0.6-0.63 Ma.
3. The Rockland Ash: Dated to ca. 0.6 Ma.
4. The Colma/Merced contact: Sarna-Wojcicki et al. (2021) identified the Paulina Lakes Ash Bed (ca. 65 ka) in the Colma Fm overlying the Merced Fm. This provides a minimum depositional age for the Colma Fm.

Additionally, the Brunhes-Matuyama reversal is ca. 773 ka. If it can be identified within the Merced Fm, that would provide another age constraint. Given these age constraints, it is possible to calculate implied sedimentation rates during different time periods (for example, to compare sedimentation rates before and after the provenance change).

Ingram and Ingle (1998) constructed two age models for the Merced Fm based on strontium isotope chemostratigraphy. They calculated an average sedimentation rate of 710 m/myr for the upper part of the section (0.1-0.8 Ma; above 720 m on the Clifton and Hunter stratigraphic column). For the lower part of the section (>0.8 Ma; below 720 m), they provide two different sedimentation rates for their two age models: 590 m/myr for Model 1 (basal age of 2.4 Ma) and 260 m/myr for Model 2 (basal age of 4.8 Ma). However, these rates are in part based on an outdated age for the Rockland Ash, and therefore require some adjustment.

## First pass at unsupervised clustering algorithm

```python
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering


def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


iris = load_iris()
X = iris.data

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

## Predicting the location of the Brunhes-Matuyama reversal based on estimates of sedimentation rate

It will be useful to have a sense of approximately where the Brunhes-Matuyama reversal is located within the Merced section, and how much that prediction is affected by different calculations of sedimentation rate.

The important horizons for these calculations are the base of the formation, the level of the provenance change (1460 m above base), and the Rockland ash bed (1580 m above base). 
- Base: 2.4-4.8 Ma ($^{87}$Sr/$^{86}$Sr; Ingram and Ingle, 1998); 3.3 Ma based on best fit simple linear age model
- Provenance change: max age of ~631 ka ($^{40}$Ar/$^{39}$Ar age for Lava Creek B ash; Matthews et al., 2015)
- Rockland ash: 609 ± 7 ka ($^{40}$Ar/$^{39}$Ar; Lanphere et al., 2004); 598 ± 13 ka (U-Pb; Coble et al., 2017)
